```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"A:\task1.csv")
df.head()
```

```python
df["Revenue"] = df["QUANTITYORDERED"] * df["PRICEEACH"]
df.head()
```

```python
total_revenue = df["Revenue"].sum()
total_orders = df["ORDERNUMBER"].nunique()
total_quantity = df["QUANTITYORDERED"].sum()

print("Total Revenue :", total_revenue)
print("Total Orders :", total_orders)
print("Total Quantity Sold :", total_quantity)
```

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"],dayfirst = True)
monthly_sales = df.groupby(df["ORDERDATE"].dt.to_period("M"))["Revenue"].sum()
monthly_sales = monthly_sales.reset_index()
monthly_sales["ORDERDATE"] = monthly_sales["ORDERDATE"].astype(str)
plt.figure(figsize=(8,4))
plt.plot(monthly_sales["ORDERDATE"], monthly_sales["Revenue"], marker='*')
plt.title("Revenue Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
top_products = df.groupby("PRODUCT")["Revenue"].sum().sort_values(ascending=False)
plt.figure(figsize=(6,3))
top_products.plot(kind="bar")
plt.title("Top Performing Products")
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.xticks(rotation=90)
plt.show()
```

```python
country_sales = df.groupby("COUNTRY")["Revenue"].sum()
plt.figure(figsize=(6,6))
country_sales.plot(kind="pie", autopct="%2.2f%%")
plt.title("Revenue by Country")
plt.ylabel("")
plt.show()
```

```python
## Conclusion
• Total revenue, orders and quantity sold were calculated as KPIs.  
• Revenue trends show monthly sales growth.  
• Top products contributing to revenue were identified.  
• Country-wise sales distribution was analyzed.  

This dashboard helps understand business performance using data analytics.
```
